```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
```

```python
def sigmoidBinary(t):
  return 1/(1+np.exp(-t))
t=np.linspace(-5,5)
  
```

```python
plt.plot(t,sigmoidBinary(t))
plt.title('binary sigmoid activation function')
plt.show()
```

Rectified Linear Unit(ReLU)

```python
def ReLU(t):
  lst=[]
  for i in t:
    if i>=0:
      lst.append(i)
    else:
      lst.append(0)
  return lst
  arr=np.linspace(-5,5)      
```

```python
plt.plot(t,ReLU(t))
plt.title('softmax Activation Function')
plt.show()
```

```python

def LeackyReLU(arr):
  lst=[]
  for i in arr:
    if i>=0:
      lst.append(i)
    else:
      lst.append(0.01*i)
  return lst

arr=np.linspace(-5,5)
plt.plot(arr,LeackyReLU(arr))
plt.title('leacky ReLU Activation Function')
plt.show()


```

```python
def softmax(t):
  return np.exp(t)/np.sum(np.exp(t))
t=np.linspace(-5,5)
plt.plot(t,softmax(t))
plt.title('softmax activation function')
plt.show()  
```
